## Conditional Statements
Conditional Statements allow you to execute code depending on whether the condition is met.

### Before Running the Example

**Predict what the code will do before executing it.**

```python
a = 7
if a == 5:
    print('a equals 5')
elif a == 6:
    print('a equals 6')
else:
    print('a is not equal to 5 or 6')
```

## Nest Conditional Statements
### Before Running the Example

**Predict what the code will do before executing it.**

```python
a = 'foo1'
b = 'baz'

if a == 'foo':
    if b == 'bar':
        print('OK this is cool!')
    else:
        print('All foo no bar')
 
```

## More Conditional Statements

```python
# %%
a = 11
if a>5 and a<10:
    print("a is between 5 and 10")
```

```python
a = 15
b = 5
c = 20
if a>b or a<c:
    print(f'a is between {b} and {c}')
    
# Notice the use of 'f' before the string declaration, this is a Formatted String Literal.
# This allows you to insert variable values into text strings
    
```

```python
a = 5
b = 10

if not a>b:
    print("a is not greater than b")
    
print('A') if not a<b else print('B')
```
